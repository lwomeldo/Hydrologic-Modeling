```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.tsa.api as sm
```

```python
df = pd.read_csv('ArrowrockReservoirElevation.csv', names=['Elevation','Y','M','D'], skiprows=1)
df['SerDates'] = pd.to_datetime(df.Y*10000+df.M*100+df.D,format='%Y%m%d')
```

```python
E = df['Elevation'].values 
Time = df['SerDates'].values 
Et   = E[1:-1] 
T   = Time[1:-1] 
plt.figure(figsize=(14,10))
plt.plot(T,Et,'o')
plt.title('Arrowrock Reservoir Elevation',fontsize=16)
plt.xlabel('$Time$ Day',fontsize=16)
plt.ylabel('$Elevation$ ft',fontsize=16)
plt.show()
```
